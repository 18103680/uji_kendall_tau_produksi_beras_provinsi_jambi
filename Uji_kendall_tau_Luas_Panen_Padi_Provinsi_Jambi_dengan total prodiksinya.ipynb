## setelah itu saya mengambil data provinsi jambi dari tahun 2018 sampai 2023

```python
df = {'Luas_Panen': [86202.68 , 69536.06 , 84772.93,64412.26, 60539.59 
                        ,61236.64], 'Produksi_Per_ton_Padi':
         [86202.68 , 69536.06 , 84772.93,298149.25, 277743.8 , 275941.45 ]}
df = pd.DataFrame(data=df)
df
```

```python
import matplotlib.pyplot as plt
# Membuat plot

plt.figure(figsize=(10, 6))
plt.scatter(df['Luas_Panen'], df['Produksi_Per_ton_Padi'], marker='o', linestyle='-', color='b')
plt.title('Luas Panen vs Produksi Per Ton Padi')
plt.xlabel('Luas Panen')
plt.ylabel('Produksi Per Ton Padi')
plt.grid(True)
plt.show()
```

```python
 
from scipy.stats import kendalltau

# Taking values from the above example in Lists
X = df['Luas_Panen']
Y = df['Produksi_Per_ton_Padi']

# Calculating Kendall Rank correlation
corr, _ = kendalltau(X, Y)
print('Kendall Rank correlation: %.5f' % corr)
```

## maka dapat disimpulkan menurut uji kendall tau korelasi luas panen padi di provinsi jambi sangat bergantung pada produksi per ton cenderung menerun sehingga pemerintah provinsi jambi harus memberikan penangulangan terhadap luas panen yang ada di provinsi jambi
